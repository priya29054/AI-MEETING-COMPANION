```python
import requests

# *** IMPORTANT: If the URL below is incorrect, update it here ***
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMSkillsNetwork-GAI-Project-C3/audio_file.mp3"

audio_file_path = "downloaded_audio.mp3"

try:
    # Send a GET request to the URL to download the file
    response = requests.get(url, allow_redirects=True, timeout=10) # Added timeout and allow_redirects

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # If successful, write the content to the specified local file path
        with open(audio_file_path, "wb") as file:
            file.write(response.content)
        print(f"File downloaded successfully to {audio_file_path}")
    else:
        # If the request failed, print an error message
        print(f"Failed to download the file. Status code: {response.status_code}, Reason: {response.reason}")
except requests.exceptions.RequestException as e:
    print(f"An error occurred during the request: {e}")
```

```python
%%writefile simple_speech2text.py
import torch
from transformers import pipeline

# Initialize the speech-to-text pipeline from Hugging Face Transformers
# This uses the "openai/whisper-tiny.en" model for automatic speech recognition.
# The `chunk_length_s` parameter specifies the chunk length in seconds.
pipe = pipeline(
    "automatic-speech-recognition",
    model="openai/whisper-tiny.en",
    chunk_length_s=30,
)

# Define the path to the audio file that needs to be transcribed
sample = 'downloaded_audio.mp3'

# Perform speech recognition on the audio file
# The `batch_size=8` parameter indicates how many chunks are processed concurrently.
# The result is stored in `prediction` with the key "text" containing the transcribed text.
prediction = pipe(sample, batch_size=8)["text"]

# Print the transcribed text to the console
print(prediction)

# Save the transcription to a text file
output_filename = "transcription.txt"
with open(output_filename, "w") as f:
    f.write(prediction)
print(f"Transcription saved to {output_filename}")
```
